```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
import sklearn
```

```python
def build_pred_fun(params,rates,seedmlp=123):
    X = np.repeat(params,len(sutil.cs),axis=0)
    X = np.hstack((X,np.tile(sutil.cs,len(params))[:,np.newaxis]))
    Y = rates.flatten()
    
    scaler = StandardScaler()
    X_train, X_test, Y_train, Y_test =\
        train_test_split(X, Y, train_size=0.80, test_size=0.20, random_state=seedmlp)
    
    scaler.fit(X_train)
    X_train = scaler.transform(X_train)
    X_test = scaler.transform(X_test)
    
    nEns = 10
    mlp_regressors = [None]*nEns
    for i in range(nEns):
        mlp_regressor  = MLPRegressor(random_state=1000*i+seedmlp,
                                  activation='relu',
                                  hidden_layer_sizes=(20, 20),
                                  max_iter=2000)
        mlp_regressor.fit(X_train, Y_train)
        mlp_regressors[i] = mlp_regressor

    def predict(X):
        predictions = np.zeros((nEns,len(X)))
        for i in range(nEns):
            predictions[i] = mlp_regressors[i].predict(X)
        return np.mean(predictions,axis=0)

    def score(X,Y):
        u = np.sum((Y - predict(X))**2,axis=0)
        v = np.sum((Y - np.mean(Y,axis=0))**2,axis=0)
        return np.mean(1 - u/v)
    
    print('Train R^2 Score : %.3f'%score(X_train, Y_train))
    print('Test  R^2 Score : %.3f'%score(X_test, Y_test))
    
    def pred(X_test):
        X_test_scaled = scaler.transform(X_test)
        Y_test=predict(X_test_scaled)
        return Y_test
          
    return pred
```

```python
rE_pred = build_pred_fun(params[good_idxs,:],rs[good_idxs,:,0])
rP_pred = build_pred_fun(params[good_idxs,:],rs[good_idxs,:,1])
rS_pred = build_pred_fun(params[good_idxs,:],rs[good_idxs,:,2])
```

```python
def pre_MD_res(params):
    rEIps = rE_pred(np.append(params,30).reshape(1,-1))[0]
    rPIps = rP_pred(np.append(params,30).reshape(1,-1))[0]
#     rSIps = rS_pred(np.append(params,30).reshape(1,-1))[0]
    rECon = rE_pred(np.append(params,50).reshape(1,-1))[0]
    rPCon = rP_pred(np.append(params,50).reshape(1,-1))[0]
#     rSCon = rS_pred(np.append(params,50).reshape(1,-1))[0]
    rEBin = rE_pred(np.append(params,80).reshape(1,-1))[0]
    rPBin = rP_pred(np.append(params,80).reshape(1,-1))[0]
#     rSBin = rS_pred(np.append(params,80).reshape(1,-1))[0]
    
    return ((rEIps-1.5)/0.5,(rECon-2.5)/0.5,(rEBin-3.5)/0.5,(rPIps-16.0)/3.0,(rPCon-20.0)/3.0)
```
